### 1. Выбрать довольно объемный текст на русском языке и провести его лемматизацию

```python
# В качестве текста выберем - https://www.litmir.me/bd/?b=55878
# Открываем текст

with open('GorkiyDetstvo.fb2', 'r') as myfile:
    text=myfile.read().replace('\n', ' ')
```

```python
# Отсечем введение в начале - его учитывать не будем

text = text[1750:]
```

```python
# Уберем html символику <p> и </p>

text = text.replace('<p>', ' ').replace('</p>', ' ')

# Выведем небольшой отрывок от того что получилось

text[0:50000]
```

```python
# Уберем номера глав

to_remove = ['I', 'II', 'III', 'IV', 'V', 'VI', 'VII', 'VIII', 'IX', 'X', 'XI', 'XII', 'XIII']


for ch in to_remove:
    if ch in text:
        text = text.replace(ch, ' ')
```

```python
# Перед тем как провести лемматизацию хорошо бы сегментировать текст на предложения, поскольку в дальнейшем, при 
# поиске биграмм, пара слов - последнее слово предложения и первое слово следующего предложения может учесться
# Сегментацию проведем с помощью NLTK 

from nltk.tokenize import sent_tokenize

sentences = sent_tokenize(text)

# Выведем небольшой отрывок получившегося 

sentences[0:5000]
```

```python
import re

# Регулярное выр-ие на все буквенные символы

regex = re.compile('[^а-яА-Я]')

# Убираем все не буквенные символы и переводим буквы в нижний регистр

new_text = []

for sentence in sentences:
    sentence = sentence.replace('ё', 'е')
    new_text.append(regex.sub(' ', sentence.lower()))

# Выведем небольшой отрывок получившегося 

new_text[0:5000]
```

```python
# Реализуем функцию, преобразующие слова в предложении в леммы

import pymorphy2
morph = pymorphy2.MorphAnalyzer()

def lemmatize(sentence):
    sentence = sentence.split()
    for i, word in enumerate(sentence):
        sentence[i] = morph.parse(word)[0].normal_form
    return sentence
```

```python
# Применим её к нашему набору предложений

lemmatize_sentences = list(map(lemmatize, new_text))
```

```python
# Итого имеем текст, разбитый на предложения, причем каждое предложение представляет собой список его 
# лемматизированных слов
```

### 2. Определить частотное слово данного текста (существительное, прилагат., глагол)

```python
# Мы храним наш текст как набор предложений (список списков), для того чтобы удобнее искать частотные слова
# схлопнем все в один список.

import itertools
full_text = list(itertools.chain.from_iterable(lemmatize_sentences))


# Теперь, посмотрим на топ-50 самых популярных словоформ 

from collections import Counter
words_dict = dict(Counter(full_text))
frequent_words = sorted(words_dict.items(), key=lambda x: x[1])[-50:]
frequent_words[::-1]
```

```python
# Теперь будем искать самое частотное слово для глагола, существительного и прилагательного 
# не будем учитывать местоимения-существительные

frequent_words = sorted(words_dict.items(), key=lambda x: x[1])

for word in reversed(frequent_words):
    if morph.parse(word[0])[0].tag.POS == 'NOUN':
        print('Самое частотное существительное: ', word)
        break
    if morph.parse(word[0])[0].tag.POS == 'VERB' or morph.parse(word[0])[0].tag.POS == 'INFN':
        print('Самое частотный глагол: ', word)
    if morph.parse(word[0])[0].tag.POS == 'ADJF' or morph.parse(word[0])[0].tag.POS == 'ADJS':
        print('Самое частотное прилагательное: ', word)
```

### 3. Подсчитать частоту встречаемости биграмм с этим словом

```python
# Для каждого из трех слов создадим список биграмм с ним

bigram_verb = []
bigram_noun = []
bigram_adjf = []


# Заполняем каждый список биграмами

for sentence in lemmatize_sentences:
    
    if ('бабушка' in sentence) and len(sentence) > 1:
        if (sentence.index('бабушка') != (len(sentence) - 1)) and (sentence.index('бабушка') != 0):
            left = sentence.index('бабушка') - 1
            right = sentence.index('бабушка') + 1
            bigram_noun.append(['бабушка', sentence[right]])
            bigram_noun.append([sentence[left], 'бабушка'])
        if sentence.index('бабушка') == (len(sentence) - 1):
            left = sentence.index('бабушка') - 1
            bigram_noun.append([sentence[left], 'бабушка'])
        if sentence.index('бабушка') == 0:
            right = sentence.index('бабушка') + 1
            bigram_noun.append(['бабушка', sentence[right]])
            
    if ('быть' in sentence) and len(sentence) > 1:
        if (sentence.index('быть') != (len(sentence) - 1)) and (sentence.index('быть') != 0):
            left = sentence.index('быть') - 1
            right = sentence.index('быть') + 1
            bigram_verb.append(['быть', sentence[right]])
            bigram_verb.append([sentence[left], 'быть'])
        if sentence.index('быть') == (len(sentence) - 1):
            left = sentence.index('быть') - 1
            bigram_verb.append([sentence[left], 'быть'])
        if sentence.index('быть') == 0:
            right = sentence.index('быть') + 1
            bigram_verb.append(['быть', sentence[right]])
    
    if ('весь' in sentence) and len(sentence) > 1:
        if (sentence.index('весь') != (len(sentence) - 1)) and (sentence.index('весь') != 0):
            left = sentence.index('весь') - 1
            right = sentence.index('весь') + 1
            bigram_adjf.append(['весь', sentence[right]])
            bigram_adjf.append([sentence[left], 'весь'])
        if sentence.index('весь') == (len(sentence) - 1):
            left = sentence.index('весь') - 1
            bigram_adjf.append([sentence[left], 'весь'])
        if sentence.index('весь') == 0:
            right = sentence.index('весь') + 1
            bigram_adjf.append(['весь', sentence[right]])
```

```python
# Выведем для каждого слова частоту биграмм

# Сначала для глагола 'быть'
new_bigram_verb = map(tuple, bigram_verb) 
verb_count = dict(Counter(new_bigram_verb))
sorted(verb_count.items(), key=lambda x: x[1])[::-1]
```

```python
# Теперь для существительного 'бабушка'

new_bigram_noun = map(tuple, bigram_noun) 
noun_count = dict(Counter(new_bigram_noun))
sorted(noun_count.items(), key=lambda x: x[1])[::-1]
```

```python
# И, наконец, для прилагательного 'весь':

new_bigram_adjf = map(tuple, bigram_adjf) 
adjf_count = dict(Counter(new_bigram_adjf))
sorted(adjf_count.items(), key=lambda x: x[1])[::-1]
```

### 4. Извлечь список из 15-20 наиболее устойчивых биграмм по мере Дайса и по мере MI

```python
# Для начала найдем все биграмы нашего текста

all_bigrams = []
```

```python
# Используем функцию nltk.bigrams библиотеки nltk

import nltk

for sentence in lemmatize_sentences:
    all_bigrams += list(nltk.bigrams(sentence))
```

```python
# Функция для Меры Dice 

def dice_score(bigram):
    
    a_1 = full_text.count(bigram[0])
    a_2 = full_text.count(bigram[1])
    a_1_2 = all_bigrams.count(bigram)
    
    return (2*a_1_2)/(a_1+a_2)
```

```python
# Функция для Меры MI 

import math
def MI_score(bigram):
    
    a_1 = full_text.count(bigram[0])
    a_2 = full_text.count(bigram[1])
    a_1_2 = all_bigrams.count(bigram)
    
    return math.log( ((len(full_text)*a_1_2)/(a_1*a_2)) , 2)
```

```python
# Для каждой биграмы вычислим меры Дайса и МI

all_bigrams_set = list(set(all_bigrams))

dice = []
MI = []

for bigram in all_bigrams_set:
    dice.append([dice_score(bigram), bigram])
    MI.append([MI_score(bigram), bigram])
```

```python
# Выведем топ20 биграм для каждой меры

# Сначала по мере Dice
from operator import itemgetter
sorted(dice, key=itemgetter(0))[-20:]
```

```python
# Теперь по мере MI

sorted(MI, key=itemgetter(0))[-20:]
```
