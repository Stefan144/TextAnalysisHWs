```python
import re

# Общие буквы кириллицы и латиницы соответственно(насколько я понял в номерах используются только большие буквы).
# Латинская буква Y не используется, она различается внешне от У, однако У может быть в номере.
common_letters = 'АВЕКМНОРСТУХABEKMHOPCTX'

#регулярное выражение для номера общего назначения
common_number = re.compile(r"^[%s]{1}[0-9]{2}[1-9]{1}[%s]{2}[1-9]?[0-9]{1}[1-9]{1}$" % 
                           (common_letters, common_letters))
#регулярное выражение для транзитного номера
transition_number = re.compile(r"^[%s]{2}[0-9]{2}[1-9]{1}[%s]{1}[1-9]?[0-9]{1}[1-9]{1}$" % 
                               (common_letters, common_letters))
```

```python
car_number = input()
```

```python
if re.match(common_number, car_number):
    print("Номер общего назначения")
elif re.match(transition_number, car_number):
    print("Транзитный номер")
else:
    print("Некорретный номер")
```
